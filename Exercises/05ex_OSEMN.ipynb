4\. load the binary file named *credit_card.dat* and convert the data into the real credit-card number.
Each line correspond to a credit card number.
Each character is composed by 6 bit (even the space) and the last 4 bit are just a padding

**hint**: use the `chr()` function to convert a number to a char

```python
n_bits = 6
cards = [] 

with open('credit_card.dat', 'rb') as f:
    for line in f:
        #Converting each line first in an integer number (base 2) and then into the corrisponding char with chr()
        card = [chr(int(line[i : i + n_bits], 2)) for i in range(0, len(line), n_bits)]
        card = card[:-1]   #Removing the last char as it is '\n'
        
        card_number = ''
        for char in card:
            card_number += char
            
        cards.append(card_number)
        
cards[:-1]  #The last 4 bit of the overall file are padding, so we discard the value we obtained

print("Credit cards:\n")
for card in cards:
    print(card)
```

7\. Load the remote file [https://www.dropbox.com/s/vkl89yce7xjdq4n/regression_generated.csv?dl=1](https://www.dropbox.com/s/vkl89yce7xjdq4n/regression_generated.csv?dl=1) with Pandas and plot a scatter plot all possible combination of the following fields:

```python
url = 'https://www.dropbox.com/s/vkl89yce7xjdq4n/regression_generated.csv?dl=1'

data = pd.read_csv(url)

#Selecting a list with the features required
basename = 'features_'
min_f = 1
max_f = 3
features = [basename + str(i) for i in range(min_f, max_f + 1)]

#Scatter plots
for i in range(len(features)):
    for j in range(i+1, len(features)):
        data.plot.scatter(features[i], features[j], figsize = (10,4))
```
